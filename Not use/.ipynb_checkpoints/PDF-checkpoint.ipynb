```python
from PyPDF2 import PdfFileReader as R
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import pdfplumber
import pandas as pd
from nltk.corpus import wordnet 
import requests
import json
#import re
import matplotlib.pyplot as plt
import seaborn as sns



from tabula import read_pdf
```

```python
#Reading resume
resume="ProviderDirectory.pdf"
with pdfplumber.open(resume) as pdf:
    no_of_pages=len(pdf.pages)
    txt1=[]
    for i in range(84,85):
        page=pdf.pages[i]
        extract=txt1.append(page.extract_text())
    print(txt1)
```

```python
#Converting multiple pages resume into single page text
sent_str = ""
for i in txt1:
    sent_str += str(i) + "\n"
sent_str = sent_str[:-1]
print(sent_str)
```

```python
#Tokenize to words
words=nltk.word_tokenize(sent_str)
words
```

```python
pos_tags= nltk.pos_tag(words)
pos_tags
```

```python
chunks= nltk.ne_chunk(pos_tags, binary=False)
for chunk in chunks:
    print(chunk)
```

```python
entities=[]
labels=[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        entities.append(''.join(c[0] for c in chunk))
        labels.append(chunk.label())
        
entites_labels = list(set(zip(entities, labels)))
entites_df = pd.DataFrame(entites_labels)
entites_df.columns=['Entites', 'Labels']
entites_df
```

```python
import nltk
from nameparser.parser import HumanName
from nltk.corpus import wordnet
```

```python
person_list = []
person_names=person_list
def get_human_names(text):
    tokens = nltk.tokenize.word_tokenize(text)
    pos = nltk.pos_tag(tokens)
    sentt = nltk.ne_chunk(pos, binary = False)

    person = []
    name = ""
    for subtree in sentt.subtrees(filter=lambda t: t.label() == 'PERSON'):
        for leaf in subtree.leaves():
            person.append(leaf[0])
        if len(person) > 1: #avoid grabbing lone surnames
            for part in person:
                name += part + ' '
            if name[:-1] not in person_list:
                person_list.append(name[:-1])
            name = ''
        person = []
#     print (person_list)
```

```python
names = get_human_names(txt1)
for person in person_list:
    person_split = person.split(" ")
    for name in person_split:
        if wordnet.synsets(name):
            if(name in person):
                person_names.remove(person)
                break

print(person_names)
```
