# 1. Top 4 teams which elected to field first after winning toss in the year 2016 and 2017.


```python
#Reading data files
df = pd.read_excel('matches.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Checking number of unique year present in the database
df['season'].value_counts()
```

```python
#Filtering out 2016 and 2017 data
year_filter = df.loc[(df['season']==2016)| (df['season']==2017)]
year_filter.season.value_counts()
```

```python
#Filtering the team who chosen fielding when they won the tose 
year_filter[year_filter['toss_decision']=='field'].head()
```

```python
#Grouping year wise toss winner team by its total number of count
grpby = year_filter.groupby('season').toss_winner.value_counts()
grpby
```

```python
#Making DataFrame
final_out = pd.DataFrame(grpby)
final_out = final_out.rename(columns={'toss_winner': 'counts'})
final_out = final_out.reset_index()
```

```python
#Segregating year wise top 4 teams
out_2016 = final_out[final_out['season']==2016]
out_2016 = out_2016[:4]
out_2017 = final_out[final_out['season']==2017]
out_2017 = out_2017[:4]
```

```python
#concating 2016 and 2017 toss winners
output1 = pd.concat([out_2016, out_2017])
output1 = output1.reset_index(drop=True)
output1
```

# Conclusion
 - On 2016 Sunrisers Hyderabad, Mumbai Indians, Delhi Daredevils and Gujarat Lions are the top 4 teams which elected to field first after winning the toss.
 - On 2017 Mumbai Indians, Kolkata Knight Riders,Royal Challengers Bangalore and Delhi Daredevils Lions are the top 4 teams which elected to field first after winning the toss.


# 2. Find the team name which has Highest Net Run Rate with respect to year.

Net Run Rate = (Total Runs Scored / Total Overs Faced) – (Total Runs Conceded /Total Overs Bowled)

```python
#Read files
df2 = pd.read_excel('deliveries.xlsx')
```

```python
df2.head()
```

```python
#Data information
df2.info()
```

```python
#Renaming columns to map the id
df2 = df2.rename(columns={'match_id':'id'})

#Merging df and df2 on id column
df3 = pd.merge(df,df2, on='id', how='left')
df3[df.columns].head(5)
```

```python
#Separating 2016 and 2017 data
df3_2016 = df3[df3['season']==2016]
df3_2017 = df3[df3['season']==2017]
```

### Year 2016

```python
# Aggregate df3_2016 by id and innings  
df4 = pd.pivot_table(df3_2016, index = ["id",'inning'], 
               values=['bowling_team', 'batting_team','over','ball', 'total_runs', 'winner','win_by_runs','win_by_wickets'],
               aggfunc={'bowling_team':lambda x: x.mode().iat[0],
                        'batting_team':lambda x: x.mode().iat[0],
                        'winner':lambda x: x.mode().iat[0],
                        'over': lambda x: x.max(),
                        'ball': np.sum,
                        'win_by_runs': np.mean,
                        'total_runs': np.sum,
                       })
#Reset index
df4=df4.reset_index()
df4.head()
```

```python
#Finding run rate for each innings
df4['run_rate']=df4['total_runs']/df4['over']
df4.head()
```

```python
#Separating Innings 1 and 2
df4_1st_innings = df4[df4['inning']==1]
df4_2nd_innings = df4[df4['inning']==2]
```

```python
#Change columns lables for each innings
colums1=[]
colums2=[]
for i in range(len(df4.columns)):
    colums1.append(df4.columns[i]+'_1')
    colums2.append(df4.columns[i]+'_2')

#Innings 1
df4_1st_innings.columns = colums1
df4_1st_innings.reset_index(inplace=True)
#Innings 2
df4_2nd_innings.columns = colums2
df4_2nd_innings.reset_index(inplace=True)
```

```python
#Concatenate both innings 
final_df_2016 = pd.concat([df4_1st_innings,df4_2nd_innings],axis=1) 
final_df_2016.head()
```

```python
#Calculating net run rate for both the innings
final_df_2016['net_run_rate'] = abs(final_df_2016['run_rate_1']-final_df_2016['run_rate_2'])
final_df_2016['net_run_rate'] = final_df_2016['net_run_rate'].astype(float)
```

```python
# Aggregate df3_2016 by unique teams  
final_df_2016 = pd.pivot_table(final_df_2016, index = ["batting_team_1"], 
               values=['net_run_rate', 'winner_1'],
               aggfunc={'winner_1':lambda x: x.mode().iat[0],
                        'net_run_rate': np.mean,
                       })
#Reset index
final_df_2016=final_df_2016.reset_index()
final_df_2016 = final_df_2016.sort_values('net_run_rate', ascending=False, ignore_index=True)
final_df_2016.head()
```

```python
#Final result
x=final_df_2016[['winner_1','net_run_rate']][:1]
x['year']=2016
```

## Year 2017

```python
# Aggregate df3_2017 by id and innings  
df5 = pd.pivot_table(df3_2017, index = ["id",'inning'], 
               values=['bowling_team', 'batting_team','over','ball', 'total_runs', 'winner','win_by_runs','win_by_wickets'],
               aggfunc={'bowling_team':lambda x: x.mode().iat[0],
                        'batting_team':lambda x: x.mode().iat[0],
                        'winner':lambda x: x.mode().iat[0],
                        'over': lambda x: x.max(),
                        'ball': np.sum,
                        'win_by_runs': np.mean,
                        'win_by_wickets':np.mean,
                        'total_runs': np.sum,
                       })
df5=df5.reset_index()
df5.head()
```

```python
#Finding run rate for each innings
df5['run_rate']=df5['total_runs']/df5['over']
df5.head()
```

```python
#Separating innings
df5_1st_innings = df5[df5['inning']==1]
df5_2nd_innings = df5[df5['inning']==2]
```

```python
#Renaming and resetting columns names and index of each innings
df5_1st_innings.columns = colums1
df5_1st_innings.reset_index(inplace=True)

df5_2nd_innings.columns = colums2
df5_2nd_innings.reset_index(inplace=True)
```

```python
#Concatenating both innings for the year 2017
final_df_2017 = pd.concat([df5_1st_innings,df5_2nd_innings],axis=1) 
final_df_2017.head()
```

```python
#Calculating net run rate for both the innings
final_df_2017['net_run_rate'] = final_df_2017['run_rate_1']-final_df_2017['run_rate_2']
final_df_2017['net_run_rate'] = final_df_2017['net_run_rate'].astype(float)
```

```python
# Aggregate df3_2017 by unique teams  
final_df_2017 = pd.pivot_table(final_df_2017, index = ["batting_team_1"], 
               values=['net_run_rate', 'winner_1'],
               aggfunc={'winner_1':lambda x: x.mode().iat[0],
                        'net_run_rate': np.mean,
                       })
#Reset index
final_df_2017=final_df_2017.reset_index()
final_df_2017 = final_df_2017.sort_values('net_run_rate', ascending=False, ignore_index=True)
final_df_2017.head()
```

```python
#Final result
y= final_df_2017[['winner_1','net_run_rate']][:1]
y['year']=2017
```

```python
output2 = pd.concat([x,y])
output2
```
